```python
#¿Qué % de valores missing exsitsen
missingV = data.isnull().sum()
missingV
```

```python
#Contamos el total de datos (celdas):
total_cells = np.product(data.shape)
total_cells
```

```python
#Contamos el total de valores missing,
#sumando todos los valores de la variable que contiene los missing para cada columna:
total_missing = missingV.sum()
total_missing
```

```python

(total_missing/total_cells)*100

Esto nos dice que en torno al 0.6 % de los valores son NA
```
